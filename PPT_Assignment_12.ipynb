```python
#solution 1
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def delete_middle(head):
    if not head or not head.next:
        return None

    slow = head
    fast = head
    prev = None

    while fast and fast.next:
        fast = fast.next.next
        prev = slow
        slow = slow.next

    if prev:
        prev.next = slow.next
    else:
        head = slow.next

    return head

def print_list(head):
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.next
    print()

# Create the linked list: 1->2->3->4->5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Original Linked List:")
print_list(head)

head = delete_middle(head)

print("Modified Linked List:")
print_list(head)
```

```python
#solution 2
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def has_cycle(head):
    if not head or not head.next:
        return False

    slow = head
    fast = head.next

    while fast and fast.next:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next

    return False


head = ListNode(1)
head.next = ListNode(3)
head.next.next = ListNode(4)
head.next.next.next = head.next

result = has_cycle(head)
print(result) 
```

```python
#solution 6
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def insert_node(head, val):
    new_node = ListNode(val)
    if not head:
        head = new_node
        return head

    tail = head
    while tail.next:
        tail = tail.next

    tail.next = new_node
    return head

def delete_nodes(head, M, N):
    if not head or M <= 0 or N <= 0:
        return head

    current = head
    previous = None

    while current:
        # Traverse M nodes
        for _ in range(M):
            if not current:
                return head
            previous = current
            current = current.next

        # Delete N nodes
        for _ in range(N):
            if not current:
                break
            current = current.next

        # Connect previous to the next node
        previous.next = current

    return head

# Create the linked list: 1->2->3->4->5->6->7->8
head = None
head = insert_node(head, 1)
head = insert_node(head, 2)
head = insert_node(head, 3)
head = insert_node(head, 4)
head = insert_node(head, 5)
head = insert_node(head, 6)
head = insert_node(head, 7)
head = insert_node(head, 8)

M = 2
N = 2

# Delete nodes
head = delete_nodes(head, M, N)

# Print the modified linked list
current = head
while current:
    print(current.val, end=" ")
    current = current.next

```

```python
#solution 7
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def insert_at_alternate_positions(first, second):
    if not second:
        return first

    if not first:
        return second

    # Initialize pointers
    first_ptr = first
    second_ptr = second

    while first_ptr and second_ptr:
        # Keep the next nodes to be connected
        first_next = first_ptr.next
        second_next = second_ptr.next

        # Insert second node after first node
        first_ptr.next = second_ptr
        second_ptr.next = first_next

        # Move pointers to next positions
        first_ptr = first_next
        second_ptr = second_next

    # Append remaining nodes of second list to first list
    if second_ptr:
        first_ptr.next = second_ptr

    # Update head of second list
    second = None

    return first


def create_linked_list(arr):
    if not arr:
        return None

    head = ListNode(arr[0])
    current = head

    for i in range(1, len(arr)):
        new_node = ListNode(arr[i])
        current.next = new_node
        current = new_node

    return head


def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()


# Example usage
first_list_values = [5, 7, 17, 13, 11]
second_list_values = [12, 10, 2, 4, 6]

first_list = create_linked_list(first_list_values)
second_list = create_linked_list(second_list_values)

print("First list before insertion:")
print_linked_list(first_list)

print("Second list before insertion:")
print_linked_list(second_list)

first_list = insert_at_alternate_positions(first_list, second_list)

print("First list after insertion:")
print_linked_list(first_list)

print("Second list after insertion (empty):")
print_linked_list(second_list)
```
