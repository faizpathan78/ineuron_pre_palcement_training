```python
#solution 1
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorder_traversal(root, values):
    if root is None:
        return
    inorder_traversal(root.left, values)
    values.append(root.value)
    inorder_traversal(root.right, values)

def assign_sorted_values(root, values):
    if root is None:
        return
    assign_sorted_values(root.left, values)
    root.value = values.pop(0)
    assign_sorted_values(root.right, values)

def convert_to_bst(root):
    values = []
    inorder_traversal(root, values)
    values.sort()
    assign_sorted_values(root, values)

def inorder_print(root):
    if root is None:
        return
    inorder_print(root.left)
    print(root.value, end=" ")
    inorder_print(root.right)

root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

print("Input:")
inorder_print(root)
print()

convert_to_bst(root)

print("Output:")
inorder_print(root)
```

```python
#solution 2
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def create_bst(values):
    if not values:
        return None

    root = TreeNode(values[0])
    for value in values[1:]:
        insert_node(root, value)

    return root

def insert_node(root, value):
    if value < root.value:
        if root.left is None:
            root.left = TreeNode(value)
        else:
            insert_node(root.left, value)
    else:
        if root.right is None:
            root.right = TreeNode(value)
        else:
            insert_node(root.right, value)

def find_distance(root, node1, node2):
    if root is None:
        return 0
    
    if node1.value < root.value and node2.value < root.value:
        return find_distance(root.left, node1, node2)
    elif node1.value > root.value and node2.value > root.value:
        return find_distance(root.right, node1, node2)
    else:
        distance1 = find_node_distance(root, node1, 0)
        distance2 = find_node_distance(root, node2, 0)
        return distance1 + distance2

def find_node_distance(root, target, distance):
    if root is None:
        return 0
    
    if root.value == target.value:
        return distance
    
    if target.value < root.value:
        return find_node_distance(root.left, target, distance + 1)
    else:
        return find_node_distance(root.right, target, distance + 1)

values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
root = create_bst(values)

node1 = root.left.right
node2 = root.right.right

distance = find_distance(root, node1, node2)
print("The distance between the two keys:", distance)
```

```python
#solution 3
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def convert_to_doubly_linked_list(root):
    if root is None:
        return None

    head, tail = convert_tree_to_list(root)

    return head

def convert_tree_to_list(node):
    if node is None:
        return None, None

    left_head, left_tail = convert_tree_to_list(node.left)
    right_head, right_tail = convert_tree_to_list(node.right)
    current_node = DoublyLinkedListNode(node.value)

    if left_head:
        left_tail.next = current_node
        current_node.prev = left_tail
    else:
        left_head = current_node

    if right_head:
        current_node.next = right_head
        right_head.prev = current_node
    else:
        right_tail = current_node

    return left_head, right_tail

root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

doubly_linked_list_head = convert_to_doubly_linked_list(root)
current = doubly_linked_list_head
while current:
    print(current.value, end=" ")
    current = current.next
```

```python
#solution 4
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None

def connect_nodes_at_same_level(root):
    if root is None:
        return root

    # Initialize the queue with the root node
    queue = [root]

    while queue:
        # Get the number of nodes at the current level
        level_size = len(queue)

        # Connect the nodes at the current level
        for i in range(level_size):
            current_node = queue.pop(0)

            # Connect to the next node in the queue
            if i < level_size - 1:
                current_node.next = queue[0]

            # Add the left and right children to the queue
            if current_node.left:
                queue.append(current_node.left)
            if current_node.right:
                queue.append(current_node.right)

    return root
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

connected_root = connect_nodes_at_same_level(root)
current = connected_root
while current:
    print(current.value, end=" → ")
    if current.next:
        print(current.next.value, end="\n")
    else:
        print("-1")
    current = current.next
```
